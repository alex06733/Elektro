```python
import matplotlib.pyplot as plt

# Daten definieren
power_values = list(range(10, 101, 10))
deviation_values = [31.95, 14.05, 8.73, 4.27, 3.15, 0.49, 0.03, -0.22, 1.02, 3.63]
error_y = [0.009] * len(deviation_values)  # Fehler der Messpunkte auf der y-Achse

# Graph erstellen
plt.figure(figsize=(10, 6))
plt.errorbar(power_values, deviation_values, yerr=error_y, fmt='o', color='blue', ecolor='red', elinewidth=1, capsize=3, label='Messpunkte mit Fehler')  # Punkte mit Fehlerbalken plotten

# Titel und Achsenbeschriftungen hinzufügen
plt.title('Abweichung der Messpunkte vom Sollwert in Abhängigkeit der Leistung', fontsize=16)
plt.xlabel('Leistung [%]', fontsize=14)
plt.ylabel('Abweichung [ml]', fontsize=14)

# Gitter und Achsenlimits setzen
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 105)
plt.ylim(min(deviation_values) - 2, max(deviation_values) + 2)

# Achsenticks anpassen
plt.xticks([x for x in range(0, 101, 10) if x != 0], fontsize=12)  # Entferne die Null von der x-Achse
plt.yticks(range(int(min(deviation_values) - 2), int(max(deviation_values) + 3), 2), fontsize=12)

# Achsenlinien entfernen und durch Pfeile ersetzen
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Pfeile zu den Achsen hinzufügen
plt.annotate('', xy=(105, 0), xytext=(0, 0), arrowprops=dict(arrowstyle='->', lw=1.5))  # Pfeil an der x-Achse
plt.annotate('', xy=(0, max(deviation_values) + 2), xytext=(0, min(deviation_values) - 2), arrowprops=dict(arrowstyle='->', lw=1.5))  # Pfeil an der y-Achse

# Legende hinzufügen
plt.legend(fontsize=12)

# Graph anzeigen
plt.tight_layout()
plt.show()
```
